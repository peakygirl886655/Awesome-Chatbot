```python
import numpy as np
import pandas as pd
```

```python
#Đọc dữ liệu vào DF
my_df = pd.read_csv("california_housing_test.csv")
```

```python
my_df
#Lưu dữ liệu ra file csv
# my_df.to_csv("my_new_data.csv")
```

```python
#Xem info ve dataframe
my_df.info()
```

```python
#Lấy thông tin theo cột và nhiều cột
my_df #hiện ra tất cả các cột
my_df['longitude'] #chỉ hiện cột longitude
my_df[['longitude','latitude']]

```

```python
#Lấy theo dòng hoặc nhiều dòng
my_df.head(5) #lấy 5 dòng đầu tiên
my_df.tail(5) #lấy 5 bản ghi cuối
my_df[0:5]    #Lấy 5 bản ghi từ 0 đến 4
```

```python
#Lấy theo điều kiện, lưu sang 1 dataframe khác
new_data_frame = my_df[(my_df.households > 110) & (my_df.population > 100)]
new_data_frame.to_csv("New_data_frame")
```

```python
#Thêm một cột mới băng latitude * 2
my_df['DoubleLatitude'] = my_df['latitude'] * 2
my_df
```

```python
#Thêm một cột Pass, nếu median_income > 5 thì pass = 1 else pass = 0
my_df['Pass'] = np.where(my_df['median_income'] > 5, 1, 0)
my_df
```

```python
#Xóa cột
# my_df = my_df.drop(columns = ['DoubleLatitude'])
# my_df
#Xóa thẳng cảng luôn: mất luôn: inplace = True
# my_df.drop(columns = ['Pass'], inplace = True)
my_df
```

```python
#Xóa hàng: muốn xóa từ hàng 3-5
my_df = my_df.drop(index = [2,3,4])
```

```python
my_df
```

```python
#pandas mạnh về thống kê và xử lý dữ liệu
my_df.sort_values(by = ['median_income']) #cột median_income tăng dần
my_df.sort_values(by = ['median_income'], ascending = False) #cột median_income tăng dần
```

```python
my_df.describe()    #thống kê
#mean: giá trị trung bình
#std: độ lệch chuẩn
#25, 50, 75 phân vị
```

```python
#trực quan hóa dữ liệu
my_df['median_income'].plot.bar(color = 'red', linestyle = '--') #ngoài ra còn có nhiều kiểu như bar, pie, ...
```

```python
#pie
group_df = my_df[['median_income', 'Pass']]
group_df
```

```python
group_df = group_df.groupby('Pass').size()
group_df
```

```python
group_df.plot.pie(subplots = True, figsize = (10,10))
```

```python
#Dùng pandas để làm sạch dữ liệu
my_df.info()
my_df.describe() #kiểm tra tính đúng đắn của dữ liệu
my_df.notna() #isna(): Kiểm tra xem có null hay không
my_df['latitude'].isna() #dropna(): xóa những dòng là na (nguy hiểm, gây mất dữ liệu)
my_df['latitude'].fillna('ABC') #điền dữ liệu
# bfill: lấy dữ liệu ngay phía trước điền vào
# ffill: lấy dữ liệu ngay sau điền vào
```
